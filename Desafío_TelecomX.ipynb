```python
#Importando bibliotecas Pandas
import pandas as pd
import requests
```

```python
# URL del archivo JSON en GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Realizar la solicitud GET
response = requests.get(url)

# Verificar que la respuesta sea exitosa
if response.status_code == 200:
    data = response.json()  # Convertir la respuesta en JSON
else:
    print("Error al obtener los datos:", response.status_code)
```

```python
# Convertir los datos JSON a un DataFrame
df = pd.DataFrame(data)

# Mostrar las primeras filas
print(df.head())

# Mostrar informaci√≥n general del DataFrame
df.info()

# Mostrar estad√≠sticas descriptivas
df.describe()
```

```python
# Expandir columnas anidadas
df_expandido = pd.concat([
    df[['customerID', 'Churn']],
    pd.json_normalize(df['customer']),
    pd.json_normalize(df['phone']),
    pd.json_normalize(df['internet']),
    pd.json_normalize(df['account'])
], axis=1)

# Ver columnas finales
print(df_expandido.columns.tolist())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Asegurar limpieza de la variable objetivo
df_expandido['evadio'] = df_expandido['evadio'].astype(str).str.strip().str.lower()
df_filtrado = df_expandido[df_expandido['evadio'].isin(['yes', 'no'])].copy()

# Variables num√©ricas de inter√©s
variables_numericas = ['Charges.Total', 'tenure', 'cuentas_diarias']
variables_numericas = [col for col in variables_numericas if col in df_filtrado.columns]

# Convertir a num√©rico (forzar)
for var in variables_numericas:
    df_filtrado[var] = pd.to_numeric(df_filtrado[var], errors='coerce')

# Gr√°ficos con estad√≠sticas
for var in variables_numericas:
    plt.figure(figsize=(8, 5))
    ax = sns.boxplot(data=df_filtrado, x='evadio', y=var)

    # Calcular estad√≠sticas por grupo (ahora s√≠ son num√©ricas)
    estadisticas = df_filtrado.groupby('evadio')[var].agg(['mean', 'median'])

    for i, categoria in enumerate(estadisticas.index):
        media = estadisticas.loc[categoria, 'mean']
        mediana = estadisticas.loc[categoria, 'median']
        plt.plot([i - 0.2, i + 0.2], [media, media], color='blue', linestyle='--', label='Media' if i == 0 else "")
        plt.plot([i - 0.2, i + 0.2], [mediana, mediana], color='red', linestyle='-', label='Mediana' if i == 0 else "")

    plt.title(f"{var} vs Evasi√≥n", fontsize=13)
    plt.xlabel("¬øEvadi√≥?")
    plt.ylabel(var)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# Informe Final: An√°lisis de Evasi√≥n de Clientes ‚Äì Telecom X

---

## üîπ Introducci√≥n

Este an√°lisis tiene como objetivo comprender el comportamiento de los clientes de Telecom X, con el fin de identificar patrones asociados a la **evasi√≥n de clientes (churn)**. La evasi√≥n ocurre cuando un cliente decide cancelar su contrato con la empresa, lo que representa una p√©rdida significativa de ingresos.

Mediante un enfoque de an√°lisis de datos, buscamos responder:  
**¬øQu√© factores est√°n m√°s relacionados con la cancelaci√≥n del servicio?**  
Este estudio pretende proporcionar insights que ayuden a tomar decisiones estrat√©gicas para **reducir la tasa de churn** y mejorar la retenci√≥n de clientes.

---

## üîπ Limpieza y Tratamiento de Datos

Los datos fueron cargados desde una API en formato JSON, estructurados en varias columnas anidadas. Se realizaron los siguientes pasos:

- **Expansi√≥n de columnas anidadas** como customer, phone, internet, y account.
- Conversi√≥n de columnas num√©ricas mal tipadas (como Charges.Total) a formato num√©rico.
- Estandarizaci√≥n de texto: todas las respuestas como "Yes" y "No" fueron convertidas a min√∫sculas y normalizadas.
- Creaci√≥n de nuevas variables √∫tiles como cuentas_diarias, estimando el gasto diario a partir de la facturaci√≥n mensual.
- Eliminaci√≥n de duplicados y tratamiento de valores nulos.

Estos pasos garantizaron que los datos estuvieran **limpios, coherentes y listos para el an√°lisis exploratorio**.

---
## üîπ An√°lisis Exploratorio

A partir del an√°lisis exploratorio realizado con visualizaciones y estad√≠sticas, se pueden destacar los siguientes hallazgos:

- La evasi√≥n (churn) est√° desbalanceada, con una mayor proporci√≥n de clientes que no evaden, seg√∫n los gr√°ficos de distribuci√≥n general.

- En las comparaciones por variables categ√≥ricas como g√©nero, m√©todo de pago y tipo de contrato, no se evidencian diferencias marcadas en los gr√°ficos que indiquen una relaci√≥n clara con la evasi√≥n. Estas variables podr√≠an requerir an√°lisis proporcionales o modelado m√°s profundo para extraer patrones m√°s concluyentes.

En cambio, las variables num√©ricas s√≠ muestran diferencias m√°s claras:

- Clientes que evadieron tienden a tener menor tiempo de permanencia en la compa√±√≠a (tenure), lo cual fue visible en los gr√°ficos de caja con l√≠neas de media y mediana.

- Tambi√©n se observ√≥ que los clientes que han gastado menos en total (Charges.Total) presentan mayor evasi√≥n, lo cual sugiere una posible relaci√≥n entre bajo compromiso financiero y mayor riesgo de cancelaci√≥n.

-La variable cuentas_diarias mostr√≥ una tendencia similar: clientes con menor gasto diario tienen m√°s probabilidad de evadir.

Estas observaciones permiten identificar perfiles de riesgo desde el comportamiento financiero y la duraci√≥n de permanencia.
