```python
#Importando bibliotecas Pandas
import pandas as pd
import requests
```

```python
# URL del archivo JSON en GitHub
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Realizar la solicitud GET
response = requests.get(url)

# Verificar que la respuesta sea exitosa
if response.status_code == 200:
    data = response.json()  # Convertir la respuesta en JSON
else:
    print("Error al obtener los datos:", response.status_code)
```

```python
# Convertir los datos JSON a un DataFrame
df = pd.DataFrame(data)

# Mostrar las primeras filas
print(df.head())

# Mostrar información general del DataFrame
df.info()

# Mostrar estadísticas descriptivas
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Asegurar limpieza de la variable objetivo
df_expandido['evadio'] = df_expandido['evadio'].astype(str).str.strip().str.lower()
df_filtrado = df_expandido[df_expandido['evadio'].isin(['yes', 'no'])].copy()

# Variables numéricas de interés
variables_numericas = ['Charges.Total', 'tenure', 'cuentas_diarias']
variables_numericas = [col for col in variables_numericas if col in df_filtrado.columns]

# Convertir a numérico (forzar)
for var in variables_numericas:
    df_filtrado[var] = pd.to_numeric(df_filtrado[var], errors='coerce')

# Gráficos con estadísticas
for var in variables_numericas:
    plt.figure(figsize=(8, 5))
    ax = sns.boxplot(data=df_filtrado, x='evadio', y=var)

    # Calcular estadísticas por grupo (ahora sí son numéricas)
    estadisticas = df_filtrado.groupby('evadio')[var].agg(['mean', 'median'])

    for i, categoria in enumerate(estadisticas.index):
        media = estadisticas.loc[categoria, 'mean']
        mediana = estadisticas.loc[categoria, 'median']
        plt.plot([i - 0.2, i + 0.2], [media, media], color='blue', linestyle='--', label='Media' if i == 0 else "")
        plt.plot([i - 0.2, i + 0.2], [mediana, mediana], color='red', linestyle='-', label='Mediana' if i == 0 else "")

    plt.title(f"{var} vs Evasión", fontsize=13)
    plt.xlabel("¿Evadió?")
    plt.ylabel(var)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# Informe Final: Análisis de Evasión de Clientes – Telecom X

---

## 🔹 Introducción

Este análisis tiene como objetivo comprender el comportamiento de los clientes de Telecom X, con el fin de identificar patrones asociados a la **evasión de clientes (churn)**. La evasión ocurre cuando un cliente decide cancelar su contrato con la empresa, lo que representa una pérdida significativa de ingresos.

Mediante un enfoque de análisis de datos, buscamos responder:  
**¿Qué factores están más relacionados con la cancelación del servicio?**  
Este estudio pretende proporcionar insights que ayuden a tomar decisiones estratégicas para **reducir la tasa de churn** y mejorar la retención de clientes.

---

## 🔹 Limpieza y Tratamiento de Datos

Los datos fueron cargados desde una API en formato JSON, estructurados en varias columnas anidadas. Se realizaron los siguientes pasos:

- **Expansión de columnas anidadas** como customer, phone, internet, y account.
- Conversión de columnas numéricas mal tipadas (como Charges.Total) a formato numérico.
- Estandarización de texto: todas las respuestas como "Yes" y "No" fueron convertidas a minúsculas y normalizadas.
- Creación de nuevas variables útiles como cuentas_diarias, estimando el gasto diario a partir de la facturación mensual.
- Eliminación de duplicados y tratamiento de valores nulos.

Estos pasos garantizaron que los datos estuvieran **limpios, coherentes y listos para el análisis exploratorio**.

---
## 🔹 Análisis Exploratorio

A partir del análisis exploratorio realizado con visualizaciones y estadísticas, se pueden destacar los siguientes hallazgos:

- La evasión (churn) está desbalanceada, con una mayor proporción de clientes que no evaden, según los gráficos de distribución general.

- En las comparaciones por variables categóricas como género, método de pago y tipo de contrato, no se evidencian diferencias marcadas en los gráficos que indiquen una relación clara con la evasión. Estas variables podrían requerir análisis proporcionales o modelado más profundo para extraer patrones más concluyentes.

En cambio, las variables numéricas sí muestran diferencias más claras:

- Clientes que evadieron tienden a tener menor tiempo de permanencia en la compañía (tenure), lo cual fue visible en los gráficos de caja con líneas de media y mediana.

- También se observó que los clientes que han gastado menos en total (Charges.Total) presentan mayor evasión, lo cual sugiere una posible relación entre bajo compromiso financiero y mayor riesgo de cancelación.

-La variable cuentas_diarias mostró una tendencia similar: clientes con menor gasto diario tienen más probabilidad de evadir.

Estas observaciones permiten identificar perfiles de riesgo desde el comportamiento financiero y la duración de permanencia.

## 🔹 Recomendaciones

- Implementar **programas de fidelización temprana** para nuevos clientes (primeros 6 meses).

- Evaluar el proceso de pago y **educar a los clientes sobre métodos más confiables**.

- Priorizar estrategias de retención en clientes de bajo gasto, como upselling o personalización de servicios.
